```python
from ScatterSim.NanoObjects import SphereNanoObject,CylinderNanoObject, PolydisperseNanoObject
# We'll import a few lattices, cubic, FCC, BCC and Diamond
from ScatterSim.LatticeObjects import Lattice, SimpleCubic, FCCLattice, BCCLattice, DiamondTwoParticleLattice
from ScatterSim.CustomLattices import BCCLatticeColor
# import the peak shape for the peaks, tunable
from ScatterSim.PeakShape import PeakShape
from ScatterSim.CompositeNanoObjects import CompositeNanoObject, OctahedronNanoObject, OctahedronCylindersNanoObject
from ScatterSim.Background import Background

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
peak = PeakShape(delta=0.01, nu=0.05) 
```

```python
# An all in one go lattice
sphere = SphereNanoObject({'radius':5})
objlist = [sphere, sphere]
positions = np.array([(0,0,0),(0.5,0.5,0.5)])
lat_sc = Lattice(objlist,lattice_positions = ['center','corner'],lattice_coordinates = positions, lattice_types=np.array([1,2]))



```

```python
# COMPLETE LATTICE DEFINITION
sphere = SphereNanoObject({'radius':5})
objlist = [sphere,sphere] #THIS MUST MATCH THE SAME LEN AS LATTICE TYPES and the Len of Coordinates

lattice_spacing_a = 50
lattice_spacing_b = None
lattice_spacing_c = None
alpha=90
beta = None
gamma = None
sigma_D = 0.01
lattice_positions = ['center','corner', 'odd']
lattice_coordinates = np.array([
    [0, 0, 0],
    [0.5,0.5,0.5]
    ])
symmetry = {
            'crystal family': 'cubic',
            'crystal system': 'cubic',
            'Bravais lattice': 'P',
            'crystal class': 'hexoctahedral',
            'point group': 'm3m',
            'space group': 'Pm3m',
            }
lattice_types=np.array([1,2])

args = [lattice_spacing_a, lattice_spacing_b, lattice_spacing_c, alpha, beta, gamma, sigma_D, lattice_positions,
       lattice_coordinates, symmetry, lattice_types]


lat_bcc_t = Lattice(objlist, *args)
```

```python
#Map intensity, and make a projection 
q = np.linspace(.4, 1, 1000)
%time lat_gen_int = lat_bcc_t.intensity(q,peak)
print('done!')

#BCC, FCC and SC
plt.figure(0, figsize=(7,5));
plt.clf()
#ROI = [1e9,1e12,0,10]
plt.title("Intensity")
plt.loglog(q, lat_gen_int,label="free")


#Visualize unit cell
length=50
extent = [-length, length, length, length]
[x_bcc,y_bcc,z_bcc]=lat_bcc_t.projections(length)

print("Finished Visualization work, lets go plot them")

plt.figure('Lattice Visual',figsize=(6,6))
plt.subplot(331)
plt.contourf(x_bcc)
plt.subplot(332)
plt.contourf(y_bcc)
plt.subplot(333)
plt.contourf(z_bcc)
```

```python
# Comparing to the BCC lattice Class that is already there for you to use


bcc_check = BCCLattice([sphere],lattice_spacing_a=50)

q = np.linspace(.4, 1, 1000)
%time bcc_check_int = bcc_check.intensity(q,peak)
print('done!')


#BCC, FCC and SC
plt.figure(0, figsize=(7,5));
plt.clf()
#ROI = [1e9,1e12,0,10]
plt.title("Intensity")
plt.loglog(q,bcc_check_int ,label="free")


#Visualize unit cell
length=50
extent = [-length, length, length, length]
[x1_bcc,y1_bcc,z1_bcc]=bcc_check.projections(length)

print("Finished Visualization work, lets go plot them")

plt.figure('Lattice check Visual',figsize=(6,6))
plt.subplot(331)
plt.contourf(x_bcc)
plt.subplot(332)
plt.contourf(y_bcc)
plt.subplot(333)
plt.contourf(z_bcc)
plt.subplot(334)
plt.contourf(x1_bcc)
plt.subplot(335)
plt.contourf(y1_bcc)
plt.subplot(336)
plt.contourf(z1_bcc)

```

```python
import numpy as np

import Part

from FreeCAD import Base 







# create the document







eta, theta, phi = 0, np.pi/4., 0

r = 3#/np.sqrt(3)

H = 28.56 # cylinder height 

eS = (38.96-28.56)/2.#*np.sqrt(3) #edgespread

dd = (H/2 + eS)/np.sqrt(2)







poslist = [

[0, 45, 90, 0, dd, -dd],

[0, 45, 0, dd, 0, -dd],

[0, 45, -90, 0, -dd, -dd],

[0, -45, 0, -dd, 0, -dd],

# now the flat part

[0, 90, -45, dd, dd, 0],

[0, 90, 45, dd, -dd, 0],

[0, 90, -45, -dd, -dd, 0],

[0, 90, 45, -dd, dd, 0],

# finally bottom part

[0, 45, 90, 0, -dd,dd],

[0, 45, 0, -dd, 0, dd],

[0, 45, -90, 0, dd, dd],

[0, -45, 0, dd, 0, dd],

]




doc = FreeCAD.newDocument()

fuse = None

for pos in poslist:

    eta, theta, phi = np.radians(pos[0]), np.radians(pos[1]), np.radians(pos[2])

    x0, y0, z0 = pos[3], pos[4], pos[5]

    # create the rotation matrix object

    myMat = Base.Matrix()

    myMat.rotateZ(eta)

    myMat.rotateY(theta)

    myMat.rotateZ(phi)

    myVec = Base.Vector(x0, y0, z0)

    # first translate to COM 




    newcyl = Part.makeCylinder(r,H)

    # move to COM

    newcyl.translate(Base.Vector(0, 0, -H/2.))

    newcyl.transformShape(myMat)

    newcyl.translate(myVec)

    if fuse is None:

        fuse = newcyl

    else:

        fuse = fuse.fuse(newcyl)




newshapeobj = doc.addObject("Part::Feature","MyShape")

newshapeobj.Shape = fuse

octahedra = fuse.copy()

doc_octa = doc
```
